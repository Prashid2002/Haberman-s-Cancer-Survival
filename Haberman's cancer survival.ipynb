```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("haberman.csv")
```

```python
data.info()
```

```python
data.columns = ["age", "op_year", "axillary_lymph_node", "survival_status"]
```

```python
data.head()
```

Survival Status (class attribute)
1 = the patient survived 5 years longer.
2 = the patient died within 5 years.

```python
data['survival_status'] = data['survival_status'].map({1:'Yes', 2:'No'}) 
```

```python
data.head()
```

```python
data.describe()
```

```python
data['survival_status'].value_counts()
```

```python
status_yes = data[data['survival_status'] == 'Yes']
status_yes.describe()
```

```python
status_no = data[data['survival_status'] == 'No']
status_no.describe()
```

## Univariant Analysis - Histogram

```python
sns.FacetGrid(data, hue = 'survival_status' ,height=5)\
.map(sns.distplot, 'age')\
.add_legend()
plt.show()
```

```python
sns.FacetGrid(data, hue = 'survival_status' ,height=5)\
.map(sns.distplot, 'op_year')\
.add_legend()
plt.show()
```

```python
sns.FacetGrid(data, hue = 'survival_status' ,height=5)\
.map(sns.distplot, 'axillary_lymph_node')\
.add_legend()
plt.show()
```

## BoxPlot

```python
sns.boxplot(x = 'survival_status', y = 'age', data = data)
plt.show()
sns.boxplot(x = 'survival_status', y = 'op_year', data = data)
plt.show()
sns.boxplot(x = 'survival_status', y = 'axillary_lymph_node', data = data)
plt.show()
```

## Pair Plot


```python
sns.set_style('whitegrid')
sns.pairplot(data, hue = 'survival_status', height=5)
plt.show()
```

## Multivariate analysis

```python
sns.jointplot(x = 'op_year', y = 'age', data = data, kind = 'kde')
plt.show()
```
